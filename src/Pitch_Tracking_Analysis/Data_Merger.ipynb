### Add Live Field to First Data Set 

```python
df = pd.read_csv('../data/feb_15_merged_data.csv')
```

```python
df.head()
```

```python
from datetime import datetime

datetime_object = datetime.strptime('Wed Feb 05 2020 19:00:00', '%a %b %d %Y %H:%M:%S')
```

```python
def label_date(row):
    print(row['date'])
    row2 = row['date'].split('GMT')[0].strip()
    datetime_object = datetime.strptime(row2,'%a %b %d %Y %H:%M:%S')
    return str(datetime_object).split(' ')[0]
```

```python
df['date_converted'] = df.apply(lambda row: label_date(row), axis = 1)
```

```python
df.head()
```

```python
live_dates = ['2020-02-01','2020-02-02','2020-02-08','2020-02-09']
```

```python
def label_live(row,live_dates):
    if row['date_converted'] in live_dates:
        return 1
    else:
        return 0
```

```python
df['live'] = df.apply(lambda row: label_live(row,live_dates), axis = 1)
```
